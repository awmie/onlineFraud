- feature engineering

````python
#encoding categorical data
categorical = ['type']
````

````python
# creating dummy variables for categorical data
categorical_dummies = pd.get_dummies(fraud_data[categorical])

categorical_dummies = categorical_dummies.astype(int)
````

````python
# concatenating the data set
fraud_data = pd.concat([fraud_data,categorical_dummies],axis=1)

fraud_data
````

````python

#remove the initial categorical columns now that we have encoded them
#use the list called categorical to delete all the initially selected columns at once
fraud_data.drop(categorical,axis=1,inplace=True)

fraud_data.drop(columns=['fraudLabel', 'nameOrig', 'nameDest'],axis=1,inplace=True)

fraud_data.head()
````

## Model selection, training and validation


 - selecting target

````python
y = fraud_data.isFraud
y
````

````python
X = fraud_data.drop('isFraud',axis=1)

X
````

importing ML algo and intiating them

````python
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
````

````python
# train test split - 80% train and 20% test

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
````

````python
# initialize the models

LR = LogisticRegression(random_state=42)
KNN = KNeighborsClassifier()
DTC = DecisionTreeClassifier(random_state=42)
RFC = RandomForestClassifier(random_state=42)
````

````python
models = [LR, KNN, DTC, RFC]
````

````python
def plot_confusion_matrix(y_test, prediction):
    cm_ = confusion_matrix(y_test, prediction)
    plt.figure(figsize= (6,4))
    sns.heatmap(cm_, cmap='coolwarm', linecolor='white', linewidths=1, annot=True, fmt='d')
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()    
````

```function``` to ```train``` **model** and ```evaluate``` **accuracy**

````python
def trainer(model, X_train, y_train, X_test, y_test):
    
    #fitting the model
    model.fit(X_train, y_train)
    
    #predict on the fitted model
    prediction = model.predict(X_test)
    
    #print evaluation metric
    print(f'\nFor {model.__class__.__name__}, Accuracy score is {accuracy_score(prediction,y_test)} \n')
    print(classification_report(y_test, prediction)) #use this later
    plot_confusion_matrix(y_test,prediction)
````

```looping``` through each model

````python
for model in models:
    trainer(model, X_train, y_train, X_test, y_test)
````

````python
# Importing the library to perform cross-validation
from sklearn.model_selection import cross_validate
import numpy as np
# Running the cross-validation on both Decision Tree and Random Forest models; specifying recall as the scoring metric
DC_scores = cross_validate(DTC, X_test, y_test, scoring='recall_macro')
RF_scores = cross_validate(RFC, X_test, y_test, scoring='recall_macro')

# Printing the means of the cross-validations for both models
print('Decision Tree Recall Cross-Validation:', np.mean(DC_scores['test_score']))
print('Random Forest Recall Cross-Validation:', np.mean(RF_scores['test_score']))
````

#### Conclusion
Upon training and evaluating our classification model, we found that the ```Decision Tree model``` performed the best by a narrow margin.

Therefore, Decision Tree performs best with recall cross-validation accuracy of `91%` which is important for our problem statement where false negative is our priority.
